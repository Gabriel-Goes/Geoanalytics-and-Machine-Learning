```python
import ternary
import pandas as pd
import matplotlib.pyplot as plt
```

## Ternary plots

## Example 1

#### importing dataframe of gamma spectrometry data 

```python
gamma_data = pd.read_csv('gamma_data.csv')
gamma_data.head()
```

### Calling the module **grid**
- This module calculate the RGB grid of the input data.

```python
# defining arguments

variables_rgb = ['KPERC', 'eTh', 'eU'] # a list with the name of the variables in the RGB order;
stretch = 3 # this argument controls the extreme values (black and white) by stretching the image;
shape = [1500, 400] # pass the shape of your data as a list [n-values in y, n-values in x].
```

```python
gamma, extent = ternary.grid(gamma_data, variables_rgb, 
                                   shape, gamma_data.E_UTM, gamma_data.N_UTM, stretch)

# the first output consists of the grid (M,N 3) used to create the RGB image.
# the second output consists of the extent (coordinates) of the espacial data we intend to plot.
```

### Calling the module plot
- This module plots the ternary map based on the grid previously calculated.

```python
ternary.plot(gamma, variables_rgb, extent, triangle = 'on')
plt.savefig('gamma_ternary_map.png', dpi = 400, bbox_inches = 'tight', transparent = False)
```
